```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import random
import matplotlib.colors as colors

```

```python
GDA2020 = gpd.read_file('SA2_2021_AUST_SHP_GDA2020/SA2_2021_AUST_GDA2020.shp', crs='EPSG:4326')
GDA2020 = GDA2020[GDA2020['GCC_NAME21'] == 'Greater Sydney']
GDA2020 = GDA2020.loc[:, ['SA2_CODE21','SA2_NAME21','geometry']]
```

```python
#random numbers will be the sigmoid values
random_numbers = [random.uniform(-1, 1) for _ in range(len(GDA2020))]
GDA2020['random_number'] = random_numbers
print(GDA2020)
```

```python

color_scale = colors.LinearSegmentedColormap.from_list(
    'measurement_cmap', ['#FFB3B3', '#FF0000'])

GDA2020.plot(column='random_number', cmap=color_scale, linewidth=0.8, edgecolor='0.8')

cax = plt.axes([0.9, 0.1, 0.03, 0.8])
sm = plt.cm.ScalarMappable(cmap=color_scale)
sm.set_array(GDA2020['random_number'])
plt.colorbar(sm, cax=cax)

plt.show()
```
