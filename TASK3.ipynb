```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from scipy import stats
```

```python
dfcode = pd.read_csv('Population.csv', usecols = ['sa2_code'])
```

```python
print(dfcode)
```

```python
data = pd.DataFrame({
    'sa2code': [],
    'percentage': []
})
failed = []

for i in range(len(dfcode)):
    code = dfcode['sa2_code'][i]
    URL = f"https://www.abs.gov.au/census/find-census-data/quickstats/2021/{code}#income-and-work"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    target_element = soup.find("th", text="In the labour force")
    if target_element is None:
        failed.append(code)
        print(f"No match found for code {code}")
    else:
        target_element = target_element.find_next("td").find_next("td")
        number = target_element.text
        float_number = float(number)
        row = {'sa2code': code, 'percentage': float_number}
        data = data.append(row, ignore_index=True)

    
    
    
```

```python
data['z_score'] = stats.zscore(data['percentage'])
data['sa2_code'] = data['sa2code'].astype('float64')
data
```

```python
name = pd.read_csv('Population.csv', usecols = ['sa2_name', 'sa2_code'])

merged_df = pd.merge(name, data, on='sa2_code')

finaldf = merged_df[['sa2_code', 'z_score']].copy()

finaldf
```
